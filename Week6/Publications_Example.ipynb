```python
import pandas as pd

df=pd.read_csv("publications_authors_first10000.csv", encoding='utf-8')
df_authors=pd.DataFrame(columns=["pmid","name","affiliation","affiliation_id","collective_name","normalized_affiliation"])


count=0
for idx, row in df.iloc[:100].iterrows():
    count +=1
    # print(f"Starting record {count}")
    pmid=str(row.pmid)
    name=row.name
    affiliation_id=str(row.affiliation)
    collective_name=str(row.collective_name)
    if "Ohio State" in affiliation_id:
        normalized_affiliation="The Ohio State University"
    else:
        normalized_affiliation=""
        
    row_for_dataframe={
    "pmid": pmid,
    "name": name,
    "affiliation": affiliation,
    "affiliation_id": affiliation_id,
    "collective_name": collective_name,
    "normalized_affiliation": normalized_affiliation
    }
    
    df_for_each_author=pd.DataFrame(row_for_dataframe, index=[0])
    df_authors=pd.concat([df_for_each_author, df_authors], axis=0, ignore_index=True)
```
